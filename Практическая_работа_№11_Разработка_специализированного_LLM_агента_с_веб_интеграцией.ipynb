```python
chat_history = []
```

```python
def financial_agent(user_input):
    global chat_history

    search_results = tavily.invoke(user_input)

    # Проверим тип и извлечём контент
    if isinstance(search_results, list):
        internet_data = "\n".join([res.get('content', '') for res in search_results[:3]])
    else:
        internet_data = str(search_results)  # на случай если вдруг что-то другое пришло

    # Промпт, учитывающий специализацию
    system_prompt = (
        "Ты выступаешь как финансовый аналитик. Используй приведённые данные из интернета "
        "и контекст диалога для формирования информированного ответа. Объясняй кратко и по делу."
    )

    # Формируем сообщения для модели
    messages = [
        {"role": "system", "content": system_prompt},
        *chat_history,
        {"role": "user", "content": f"{user_input}\n\nВот данные из поиска:\n{internet_data}"}
    ]

    # Вызов LLM через Ollama
    response = ollama.chat(model=MODEL, messages=messages)
    reply = response['message']['content']

    # Обновляем историю
    chat_history.append({"role": "user", "content": user_input})
    chat_history.append({"role": "assistant", "content": reply})

    return reply
```

```python
def save_chat_history(filename):
    with open(filename, "w", encoding="utf-8") as f:
        for msg in chat_history:
            role = "Пользователь" if msg["role"] == "user" else "Агент"
            f.write(f"{role}: {msg['content']}\n\n")
    return f"История сохранена в файл: {filename}"
```
